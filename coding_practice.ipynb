```python
#write a code to find palindrome or not

class palindrome:
    def find(self,word):
        word=str(word)
        reverse=word[::-1]

        if word==reverse:
            return "palindrome"
        return "not a palindrome"
    
ex=palindrome()
print(ex.find(1541))
```

```python
#Sum of first N natural numbers

class natural:
    def find(self,end):
        sum=0
        for i in range(end+1):
            sum+=i
        print(sum)

ex=natural()
ex.find(3)
```

```python
#sum of first N natural numbers using recursion

def getsum(num):
    if num==1:
        return 1
    return num+getsum(num-1)

num=5
print(getsum(num))

# getSum(5) = 5 + getSum(4) = 15
# getSum(4) = 4 + getSum(3) = 10
# getSum(3) = 3 + getSum(2) = 6
# getSum(2) = 2 + getSum(1) = 3
# getSum(1) = 1

```

```python
#Find the prime numbers between 1 to 100

def is_prime(n):
    if n<=1:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

start=10
end=50

for num in range(start,end+1):
    if is_prime(num):
        print(num)

        
```
