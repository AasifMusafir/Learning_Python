```python
#Variable Length Arguments

#Positional and Keyword Argument

def print_numbers(*args):
  for numbers in args:
    print(numbers)
```

```python
print_numbers(1,2,3,4,5,"args")
```

```python
##Keywords Arguments
#In keyword Argument all the parameters will be in key value pair

def print_detail(**kwargs):
  for key,value in kwargs.items():
    print(f"{key}:{value}")

```

```python
print_detail(name="Asif",age="23",country="India")
```

```python
print_detail(name="Arif",age="23",country="India")
```

```python
#Implementing both keyword and Positional Argument

def print_details(*args,**kwargs):
  for num in args:
    print(f"Positional argument is :{num}")

  for key,value in kwargs.items():
    print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"args",name="Asif",age="23",country="India")
```

```python
 #Temperature Conversion

def convert(temp,unit):
  """This function converts temperature between Celsius and farhenit"""
  if unit=="C":
    return temp *9/5+32 #converting the celsius to farhenit
  elif unit=="F":
    return (temp-32)*5/9 #Farhrenit to celsius
  else:
    return None


print(convert(25,"C"))
print(convert(77,"F"))
```

```python
#Password strength checker
def password_checker(password):
  if len(password)<8:
    return False
  if not any(char.isdigit() for char in password):
    return False
  if not any(char.islower() for char in password):
    return False
  if not any(char.isupper() for char in password):
    return False
  if not any(char in "!@#$%^&*" for char in password):
    return False
  return True
```

```python
password_checker("Asif@122")
```

```python
#Calculating the total cost of items in a cart
def calculate(cart):
  total_cost=0
  for item in cart:
    total_cost+=item["price"]*item["quantity"]

  return total_cost
```

```python
cart=[{"name":"apple","price":1,"quantity":4},
      {"name":"banana","price":50,"quantity":1},
      {"name":"kiwi","price":2,"quantity":1}]
```

```python
#calling the function
total_cost=calculate(cart)
print(total_cost)
```

```python
def is_palindrome(s):
  s=s.lower().replace(" ","")
  return s==s[::-1]

is_palindrome("GADA g")
```

```python
#Calculate a factorial of a number using the recursion
def factorial(n):
  if n==0:
    return 1
  else:
    return n*factorial(n-1)

print(factorial(3))
```

```python
#A function to read file and count the frequency of each word
def count_frequency(file_path):
  word_count={}
  with open(file_path,"r") as file:
    for line in file:
      words=line.split()
      for word in words:
        word=word.lower().strip('.,?;:"\'')
        word_count[word]=word_count.get(word,0)+1
  return word_count

file_path="/content/Hello.txt"
word_count= count_frequency(file_path)
print(word_count)
```
