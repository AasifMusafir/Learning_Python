#Polymorphism

**Polymorphism means "many forms". It refers to the ability of an entity (like a function or object) to perform different actions based on the context.**

**Technically, in Python, polymorphism allows same method, function or operator to behave differently depending on
object it is working with. This makes code more flexible and reusable.**

#Method Overridding

**Method overridding allows a child class to provide specific implementation of a method that is already defined in its parent class**

```python
#Base class
class Animal:
  def speak(self):
    return "Sound of an Animal"

class Dog:
  def speak(self):
    return "Woof"

#child class
class cat(Animal):
  def speak(self):
    return "Meow"

#Function that demonstrate polymorphism
def Animal_speak(Animal):
  print(Animal.speak())

dog=Dog()
print(dog.speak())
Animal_speak(Dog())
```

```python
#Polymorphism with Functions and Methods
class shape:
  def area(self):
    return "the area of the shape"

class rectangle(shape):
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def area(self):
    return self.width*self.height

class circle(shape):
  def __init__(self,radius):
    self.radius=radius

  def area(self):
    return 2*self.radius*self.radius

#Function that demonstrates Polymorphism
def print_area(shape):
  print(f"the area is {shape.area()}")

Rectangle=rectangle(2,3)
print_area(Rectangle)
Circle=circle(2)
print_area(Circle)
```

#Polymorphism using an Abstract Class

**An abstract class defines a required interface that child classes must implement.
Polymorphism still happens at runtime, but now the contract is explicit and enforced.**

```python
#Polymorphism with Abstract Base Class
from abc import ABC,abstractmethod

#define an abstract class
class Vehicle(ABC):
  @abstractmethod
  def start_engine(self):
    pass

class Car(Vehicle):
  def start_engine(self):
    return "Car engine is starting"

class Motor(Vehicle):
  def start_engine(self):
    return "Motor Vehicle engine is starting"

#implementing a function that demonstrates Polymorphism

def start_vehicle(Vehicle):
  print(f"{Vehicle.start_engine()}")

#creating an objects
car1=Car()
Motor1=Motor()
start_vehicle(car1)
start_vehicle(Motor1)
```
