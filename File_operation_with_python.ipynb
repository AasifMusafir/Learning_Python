#FILE OPERATIONS-Read and write Files

**File Handling is a crucial part of any programming language>Python provides built-in-functions and methods to read from and write to files,both text and binary.This notebook covers the basics of file handling,including reading and writing text files and binary files**

```python
#read a Whole File

with open("/content/Hello.txt","r") as file:
  content=file.read()
  print(content)
```

```python
#Read a file line by line
with open("/content/Hello.txt","r") as file:
  for line in file:
    print(line.strip()) #.strip removes the newline character
```

```python
#Writing a file(Overwriting)

with open("/content/Hello.txt","w") as file:
  file.write("Hello World\n")
  file.write("this is a new line.")
```

```python
#writng a file(without overwriting)
with open("/content/Hello.txt","a") as file:
  file.write("\nthere is a append now\n")
```

```python
#writing a list of lines to a file
lines=["First line\n","second line\n","third line\n"]
with open("/content/Hello.txt","a") as file:
  file.writelines(lines)
```

```python
##Binary Files

data=b'\x00\x01\x02\x03\x04'
with open("/content/binary.bin","wb") as file:
  file.write(data)
```

```python
##Binary Files
with open("/content/binary.bin","rb") as file:
  content=file.read()
  print(content)
```

```python
#Read the content from a source file and write it to a destination file
with open("/content/Hello.txt","r") as source_file:
  content=source_file.read()

with open("destination.txt","w") as destination_file:
  destination_file.write(content)
```

```python
#Read a text file amd count the number of lines,words and characters.
def count_text_file(file_path):
  with open(file_path,"r") as file:
    lines=file.readlines()
    line_count=len(lines)
    word_count=sum(len(line.split()) for line in lines)
    char_count=sum(len(line) for line in lines)

  return line_count,word_count,char_count
file_path="/content/Hello.txt"
lines,words,characters=count_text_file(file_path)
print(f"Lines:{lines},Words:{words},Characters:{characters}")
```

**The w+ mode in Python is used to open a file for both reading and writing.If the file does not exist,it will be created.If the file exists,its content is truncated(i.e file is overwritten)**

```python
#writing a file and then reading a file

with open("/content/Hello.txt","w+") as file:
  file.write("Hello world\n")
  file.write("This is a new lline \n")

  ##Move the file cursor to the begining
  file.seek(0)

  ##read the content of the file
  content=file.read()
  print(content)
```
