#Example of a class

```python
#Modelling a bank account
class BankAccount:
  def __init__(self,user,balance=0):
    self.user=user
    self.balance=balance

  def deposit(self,amount):
    self.balance+=amount
    print(f"The updated balance is {self.balance}")

  def withdraw(self,amount):
    if amount>self.balance:
      print("Insufficient Funds")
    else:

      self.balance-=amount
      print(f"The updated balance is {self.balance}")

  def status(self):
    print(f"the balance is {self.balance}")
```

```python
User=BankAccount("asif",550)
```

```python
User.deposit(500)
```

```python
User.status()
```

```python
User.withdraw(20000)
```

#**INHERITANCE**

**Inheritance is a fundamental concept in Object Oriented Programming(OOPS) that allows a class to inherit attributes and methods from one class to another class.This notebook covers single inheritance, multiple inheritance**

```python
#Creating a class
#Parent class
class Car:
  def __init__(self,model,year):
    self.model=model
    self.year=year

  def drive(self):
    print(f"The person is driving {self.model} of {self.year}")

car1=Car("hyundia",2023)
car1.drive()
```

```python
#Inheriting the attributes and methods from the parent classs
class tesla(Car):
  def __init__(self,model,year,auto):
    super().__init__(model,year)
    self.auto=auto

  def detail(self):
    print(f"the {self.model} being autonomous is {self.auto}")

car2=tesla("tesla",2024,"True")

car2.detail()
car2.drive()
```

#Multiple Inheritance
**When a class inherits from more than a base class**

```python
#multiple Inheritance

#Creating a base class 1
class Animal:
  def __init__(self,name):
    self.name=name

#Creating a base class 2
class owner:
  def __init__(self,petowner):
    self.petowner=petowner

#Inheriting class from both base class(Multiple class)
class speaker(Animal,owner):
  def __init__(self,name,petowner):
    Animal.__init__(self,name)
    owner.__init__(self,petowner)

  def speak(self):
    print(f"The Animal is {self.name} and the owner is {self.petowner}")

detail1=speaker("Dog","Asif")

detail1.speak()
```
